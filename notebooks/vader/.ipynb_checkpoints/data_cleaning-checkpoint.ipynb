```python
import os

data = []

for root, dirs, files in os.walk("../../hutto_ICWSM_2014/results/"):
    for f in files:
        if f.endswith(".txt"):
            for l in open(os.path.join(root, f), "r"):
                l = l.split("\t")[1:]
                if (len(l) == 2):
                    l[0] = float(l[0]) / 4
                    if l[1].endswith("\n"):
                        l[1] = l[1][:-1]
                    data.append(l)

                
print(len(data))
res = open("../../hutto_ICWSM_2014/results/data_modified.txt", "w")
res.write(str(data))
res.close()
```

```python
from vaderSentiment import sentiment

for d in data:
    s = sentiment(d[1])
    if (s["pos"] > s["neg"]):
        print("pos")
    else
    
    print(s)
```
