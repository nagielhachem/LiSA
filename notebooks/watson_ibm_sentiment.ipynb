```python
import json
from watson_developer_cloud import NaturalLanguageUnderstandingV1
import watson_developer_cloud.natural_language_understanding.features.v1 \
    as Features
    
natural_language_understanding = NaturalLanguageUnderstandingV1(
    username="6ee07262-c6b4-490a-a183-7ff94bb43211",
    password="wECprNXKfN3J",
    version="2017-09-21")

path = "../data/michigan/training.txt"

def getDataLines(file_path):
    f = open(file_path, "r")
    lines = f.readlines()
    f.close
    return lines

def get_sentiment(t):
    response = natural_language_understanding.analyze(
        text=t,
        features=[
            Features.Sentiment(
                document=True
            )   
        ],
        language='en'
    )
    if ('sentiment' in response):
        sentiment = response['sentiment']
        if ('document' in sentiment):
            document = sentiment['document']
            if ('score' in document):
                return document['score']
    return 0.0 # don't know

lines = getDataLines(path)

count  = 0
acc = 0
dont_know = 0

for l in lines:
    l = l.split("\t")
    l[1] = l[1][:-1]
    feeling = get_sentiment(l[1])
    if (feeling == 0):
        dont_know = dont_know + 1
    else:
        label = int(l[0])
        if ((feeling > 0 and label == 1) or (feeling < 0 and label == 0)):
            acc = acc + 1
    count = count + 1
    if (count % 10 == 0) :
        print(count)

print("Neutral (don't know ?) : " + str(dont_know) + " ("  + str(float(dont_know) / float(count)) + "%)")
print("Correct                : " + str(acc) + " (" + str(float(acc) / float(count)) + "%)")
```
