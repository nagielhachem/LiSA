```python
from random import random

N = 100

for f in files:
    print(f)
    f = open(f, "r")
    lines = json.loads(f.readlines()[0])
    
    count  = 0
    acc = 0
    dont_know = 0
    for l in lines:
        if random() < N / len(lines):
            feeling = get_sentiment(l[1])
            if (feeling == 0):
                dont_know = dont_know + 1
            else:
                label = l[0]
                if ((feeling > 0 and label == 1) or (feeling < 0 and label == 0)):
                    acc = acc + 1
            count = count + 1
            if (count % 10 == 0) :
                print(count)

    print("Neutral (don't know ?) : " + str(dont_know) + " ("  + str(100 * float(dont_know) / float(count)) + "%)")
    print("Correct                : " + str(acc) + " (" + str(100 * float(acc) / float(count)) + "%)")
```
