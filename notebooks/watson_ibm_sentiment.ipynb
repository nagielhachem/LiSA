```python
from random import random

N = 100

for f in files:
    print(f)
    f = open(f, "r")
    lines = json.loads(f.readlines()[0])
    
    count  = 0
    tp = 0
    fp = 0
    tn = 0
    fn = 0

    dont_know = 0
    for l in lines:
        if random() < N / len(lines):
            feeling = get_sentiment(l[1])
            if (feeling == 0):
                dont_know = dont_know + 1
            else:
                label = l[0]
                if (feeling > 0):
                    if (label == 1):
                        tp += 1
                    else:
                        fp += 1
                else:
                    if (label == 0):
                        tn += 1
                    else:
                        fn += 1
            count = count + 1
            #if (count % 10 == 0) :
            #    print(count)

    print("Neutral (don't know ?) : " + str(dont_know) + " ("  + str(100 * float(dont_know) / float(count)) + "%)")
    if (tp + tn + fp + fn):
        accuracy = (tp + tn) / (tp + tn + fp + fn)
    else:
        accuracy = 0
    print("Accuracy               : " + str(100 * accuracy) + "%")
    if (tp + fp):
        precision = tp / (tp + fp)
    else:
        precision = 0
    print("Precision              : " + str(100 * precision) + "%")
    if (tp + fn):
        recall = tp / (tp + fn)
    else:
        recall = 0
    print("Recall                 : " + str(100 * recall) + "%")
    if (recall and precision):
        f1 = 2 / ((1/recall) + (1/precision))
    else:
        f1 = 0
    print("F1 score               : " + str(100 * f1) + "%")
```
